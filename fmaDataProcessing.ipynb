```python
import pydub
import soundfile as sf
from Datasets.fma.fma import utils

"""
Fma dataset reading
"""

tracks = utils.load('Datasets/fma/fma/data/fma_metadata/tracks.csv')
features = utils.load('Datasets/fma/fma/data/fma_metadata/features.csv')
small = tracks['set', 'subset'] <= 'small'

electronic = tracks['track', 'genre_top'] == 'Instrumental'
electronic_ids = tracks.loc[small & electronic, ('track', 'genre_top')].keys().tolist()

experimental = tracks['track', 'genre_top'] == 'Experimental'
experimental_ids = tracks.loc[small & experimental, ('track', 'genre_top')].keys().tolist()
experimental_ids.remove(133297) # Removing corrupted audio

folk = tracks['track', 'genre_top'] == 'Folk'
folk_ids = tracks.loc[small & folk, ('track', 'genre_top')].keys().tolist()

hiphop = tracks['track', 'genre_top'] == 'Hip-Hop'
hiphop_ids = tracks.loc[small & hiphop, ('track', 'genre_top')].keys().tolist()

instrumental = tracks['track', 'genre_top'] == 'Instrumental'
instrumental_ids = tracks.loc[small & instrumental, ('track', 'genre_top')].keys().tolist()

international = tracks['track', 'genre_top'] == 'Instrumental'
international_ids = tracks.loc[small & international, ('track', 'genre_top')].keys().tolist()

pop = tracks['track', 'genre_top'] == 'Pop'
pop_ids = tracks.loc[small & pop, ('track', 'genre_top')].keys().tolist()

rock = tracks['track', 'genre_top'] == 'Rock'
rock_ids = tracks.loc[small & rock, ('track', 'genre_top')].keys().tolist()
rock_ids.remove(108925) # Removing corrupted audio
```

```python
import os

AUDIO_DIR = 'Datasets\\fma\\fma\\data\\fma_small\\'

def read_fma_audio(AUDIO_DIR, track_id):
    filename = AUDIO_DIR + str(track_id).zfill(6) + '.mp3'

    data = sf.read(filename)
    
    return data
    
#     return librosa.load(filename, sr=None, mono=True)

def get_fma_audio_name(AUDIO_DIR, track_id):
    return AUDIO_DIR + str(track_id).zfill(6) + '.mp3'

def cut_fma_audio(AUDIO_DIR, track_id, genre, segment_duration):
    file_name = get_fma_audio_name(AUDIO_DIR, track_id)
    t1 = 1*1000
    t2 = segment_duration*1000
    audio = pydub.AudioSegment.from_file(file_name)
    
    new_audio = audio[t1:t2]
    print(new_audio)
    new_file_name = file_name.split('\\')[-1].split('.')[0]
    print(new_file_name)
    song_out  =  os.path.join(r"C:\Users\alvar\github\gitTfgPy\TFGNotebooks\Datasets\fma\small_dataset_reduced_10s", f'{genre}', f'{new_file_name}')
    print(song_out)
    new_audio.export(song_out, format="wav")
    return new_audio

audios_electronic = [
    cut_fma_audio(AUDIO_DIR, track_id, 'electronic', 10)[0] for track_id in electronic_ids
]
    
audios_experimental = [
    cut_fma_audio(AUDIO_DIR, track_id, 'experimental', 10)[0] for track_id in experimental_ids
] 

audios_folk = [
    cut_fma_audio(AUDIO_DIR, track_id, 'folk', 10)[0] for track_id in folk_ids
]

audios_hiphop = [
    cut_fma_audio(AUDIO_DIR, track_id, 'hiphop', 10)[0] for track_id in hiphop_ids
] 
    
audios_instrumental = [
    cut_fma_audio(AUDIO_DIR, track_id, 'instrumental', 10)[0] for track_id in instrumental_ids
]

audios_international = [
    cut_fma_audio(AUDIO_DIR, track_id, 'international', 10)[0] for track_id in international_ids
] 

audios_pop = [
    cut_fma_audio(AUDIO_DIR, track_id, 'pop', 10)[0] for track_id in pop_ids
] 

audios_rock = [
    cut_fma_audio(AUDIO_DIR, track_id, 'rock', 10)[0] for track_id in rock_ids
] 
    
```

```python
from pathlib import Path
from os import remove
import librosa


"""
Clean audio files of possible empty files
"""

def read_and_clean_empty_audios(path):
    audio = librosa.load(path, 44100)[0]
    if audio.size == 0:
        remove(path)
        print(path, "removed")

for p in Path().glob('Datasets/fma/small_dataset_reduced_10s/hiphop/*'):
    read_and_clean_empty_audios(p)

for p in Path().glob('Datasets/fma/small_dataset_reduced_10s/instrumental/*'):
    read_and_clean_empty_audios(p)

for p in Path().glob('Datasets/fma/small_dataset_reduced_10s/international/*'):
    read_and_clean_empty_audios(p)
    
for p in Path().glob('Datasets/fma/small_dataset_reduced_10s/pop/*'):
    read_and_clean_empty_audios(p)
    
for p in Path().glob('Datasets/fma/small_dataset_reduced_10s/rock/*'):
    read_and_clean_empty_audios(p)
```
